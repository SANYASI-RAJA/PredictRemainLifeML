```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('../data/train_FD001.txt', sep='\s+', header=None, engine='python')
print(df1)
```

```python
df1.describe()
```

```python

null_counts = df1.isna().sum()
print(null_counts)
# Print status for every column
for col in df1.columns:
    count = null_counts[col]
    if count > 0:
        print(f"Column '{col}' has {count} null values.")
    else:
        print(f"Column '{col}' has no null values.")
```

```python
for col, dtype in df1.dtypes.items():
    print(f"Column '{col}' has data type '{dtype}'")
```

```python
columns = ['unit_number','time_in_cycles','setting_1','setting_2','TRA','T2','T24','T30','T50','P2','P15','P30','Nf',
           'Nc','epr','Ps30','phi','NRf','NRc','BPR','farB','htBleed','Nf_dmd','PCNfR_dmd','W31','W32' ]
df1.columns=columns
print(df1)
```

```python
# Plot histograms for each column
for col in df1.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df1[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#delete columns with constant values â€‹â€‹that do not carry information about the state of the unit
df1.drop(columns=['Nf_dmd','PCNfR_dmd','P2','T2','TRA','farB','epr'],inplace=True)
print(df1)
```
